```python
profiles=concat_company_profiles(ticket_collection, clean_html=True)
profiles
```

```python
df2=stock.listing.all_symbols()
df2=df2.rename(columns={'organ_name':'company_name'})
df2
```

```python
merged_df = pd.merge(profiles, df2, on='company_name', how='left')
last_column = merged_df.columns[-1]
columns = [last_column] + [col for col in profiles.columns if col != last_column]

profiles=merged_df[columns]
profiles['description'] = profiles.apply(row_to_text, axis=1).tolist()
profiles
```

```python
profiles.to_excel(f'{data_dir}/profiles_vie.xlsx', index=None, engine='xlsxwriter')
```
